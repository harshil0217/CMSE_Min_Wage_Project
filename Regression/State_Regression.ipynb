```python
import pandas as pd
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import RobustScaler
```

```python
state_econ = pd.read_csv("../Data/state_econ.csv")
```

```python
state_econ
```

```python
state_econ = state_econ.set_index(["State", "Year"])
```

```python
state_econ
```

```python
corr_coef = state_econ.corr()
corr_coef
```

```python
scaler = RobustScaler()
scaler.set_output(transform = "pandas")
scaler.fit_transform(state_econ)
```

```python
model = sm.OLS(state_econ["gini"], sm.add_constant(state_econ[state_econ.columns[1:]]))
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
"""
The following features are being dropped for not meeting the threshold for statistical significance (we chose p<0.15 because
many of our correlations are not strong in the first place)
"""

state_econ = state_econ.drop(["unemployment", "GDP (in millions USD)", "GDP per Capita", "Residents with Bachelor's Degree or Higher", "Union Membership %"], axis = 1)
```

```python
state_econ
```
