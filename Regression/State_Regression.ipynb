```python
import pandas as pd
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import RobustScaler
```

```python
state_econ = pd.read_csv("../Data/state_econ.csv")
```

```python
state_econ
```

```python
state_econ = state_econ.set_index(["State", "Year"])
```

```python
state_econ
```

```python
corr_coef = state_econ.corr()
corr_coef
```

```python
#Removed features that are highly colinear with other features and with each other
state_econ = state_econ.drop(["Population", "GDP (in millions USD)"], axis = 1)
```

```python
scaler = RobustScaler()
scaler.set_output(transform = "pandas")
scaler.fit_transform(state_econ)
```

```python
model = sm.OLS(state_econ["gini"], sm.add_constant(state_econ[state_econ.columns[1:]]))
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
"""
The following features are being dropped for not meeting the threshold for statistical significance or having a coefficient
that is essentially zero (suggesting the model regards that variable as irrelevant)

Unfortunately, this means our regression only uses 3 of initial variables to predict gini coefficient
"""

state_econ = state_econ.drop(["unemployment", "GDP per Capita", "Union Membership %",
                              "Residents with Bachelor's Degree or Higher"], axis = 1)
```

```python
state_econ
```

```python
gini_model = sm.OLS(state_econ["gini"], sm.add_constant(state_econ[state_econ.columns[1:]]))
```

```python
results = gini_model.fit()
```
