```python
def merge_data(path, state_econ):
    """Merge additional data into the state_econ DataFrame.

    Takes a CSV dataset located at the path and merges it into the 
    state_econ DataFrame based on shared State and Year index values.

    Args:
        path (str): Filepath of the CSV dataset to merge.
        state_econ (DataFrame): DataFrame with State and Year index to merge data into.

    Returns: 
        DataFrame: Merged state_econ DataFrame with additional columns from CSV.
        
    Anthropic.(2023). Claude. Accessed from https://claude.ai/chat on Nov 26, 2023. 
    Prompt: {write a docstring for the following function}
    """
    data = pd.read_csv(path,dtype = {"Year": str} )
    data = data[data["State"].isin(gini["State"])]
    data = data.set_index(["State", "Year"])
    state_econ = pd.concat([state_econ, data], axis = 1)
    return state_econ
    
```

```python
state_econ = merge_data("Unemployment Data/unemployment_data.csv", state_econ)
```
