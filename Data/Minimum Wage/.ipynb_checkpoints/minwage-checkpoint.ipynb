```python
import pandas as pd
```

```python
#data source : https://www.dol.gov/agencies/whd/state/minimum-wage/history
```

```python
one = pd.read_csv("minwage2007-2013.csv", skiprows = 1)
```

```python
one = one.drop(["2007", "2008", "2009"], axis = 1)
```

```python
one
```

```python
one = pd.melt(one, id_vars = "State or other jurisdiction", var_name = "Year", value_name = "min_wage")
```

```python
one["min_wage"] = one["min_wage"].str.extract("(\d+\.\d+)")
```

```python
one
```

```python
two = pd.read_csv("minwage2014-2019.csv", skiprows = 1)
```

```python
two = pd.melt(two, id_vars = "State or other jurisdiction", var_name = "Year", value_name = "min_wage")
```

```python
two["min_wage"] = two["min_wage"].str.extract("(\d+\.\d+)")
```

```python
two
```

```python
three = pd.read_csv("minwage2020-2022.csv", skiprows = 1)
three = pd.melt(three, id_vars = "State or other jurisdiction", var_name = "Year", value_name = "min_wage")
three["min_wage"] = three["min_wage"].str.extract("(\d+\.\d+)")
```

```python
three
```

```python
minwage = pd.concat([one, two, three])
```

```python
minwage
```

```python
minwage["min_wage"] = pd.to_numeric(minwage["min_wage"])
```

```python
minwage = minwage.fillna(7.25)
```

```python
minwage
```

```python
minwage.to_csv("../min_wage.csv", index = False)
```
