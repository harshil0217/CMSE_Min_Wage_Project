```python
"""
The following code cleans and reformats unemployment data obtained from a kaggle dataset curated by user Justin Oh

https://www.kaggle.com/datasets/justin2028/unemployment-in-america-per-us-state/data

"""
```

```python
import pandas as pd
```

```python
unemployment = pd.read_csv("Unemployment in America Per US State.csv")
```

```python
unemployment
```

```python
unemployment = unemployment.drop(unemployment.columns[4:10], axis = 1)
```

```python
unemployment.drop(["FIPS Code"], axis = 1)
```

```python
unemployment = unemployment.groupby(["State/Area", "Year"]).mean()
unemployment
```

```python
unemployment = unemployment.drop(["Month"], axis =1)
```

```python
unemployment = unemployment.reset_index()
```

```python
unemployment = unemployment.drop(["FIPS Code"], axis = 1)
```

```python
unemployment = unemployment.rename(columns = {"State/Area":"State", "Percent (%) of Labor Force Unemployed in State/Area": "Unemployment"})
```

```python
unemployment
```

```python
unemployment = unemployment[unemployment["Year"] >= 2010]
```

```python
unemployment
```

```python
unemployment = unemployment[unemployment["Year"] != 2020]
```

```python
unemployment = unemployment.pivot(index = "State", columns = ["Year"], values = "Unemployment")
```

```python
unemployment
```

```python
unemployment = unemployment.reset_index()
```

```python
unemployment = pd.melt(unemployment, id_vars = "State", var_name = "Year", value_name = "unemployment")
unemployment
```

```python
unemployment.to_csv("unemployment_data.csv", index = False)
```
