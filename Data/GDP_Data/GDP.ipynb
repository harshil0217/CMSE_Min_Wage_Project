```python
"""
The following code cleans and reformats a dataset found here on kaggle that documents various state level macroeconomic 
indicators from years 1997-2022
(https://www.kaggle.com/datasets/davidbroberts/us-gdp-by-state-19972020)

The code extracts GDP data from every state for years 2010-2019, 2021-2022, excluding 2020 because 2020 data is not
available for several other variables in the dataset
"""
```

```python
import pandas as pd
```

```python
gdp = pd.read_csv("SAGDP2N__ALL_AREAS_1997_2020.csv")
```

```python
gdp
```

```python
gdp = gdp[gdp["LineCode"] == 1.0]
```

```python
gdp
```

```python
gdp = gdp.drop(["GeoFIPS", "Region", "TableName", "LineCode", "IndustryClassification", "Description", "Unit"], axis = 1)
```

```python
gdp
```

```python
gdp = gdp[~gdp["GeoName"].isin(["United States *", "District of Columbia", "New England", "Mideast", "Plains", "Southeast", "Southwest", "Rocky Mountain", "Far West", "Great Lakes"])]
```

```python
gdp
```

```python
gdp = pd.melt(gdp, id_vars = ["GeoName"], var_name = "Year", value_name = "GDP (in millions USD)")
```

```python
gdp
```

```python
gdp = gdp[gdp["Year"].isin(["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2021", "2022"])]
```

```python
gdp
```

```python
gdp = gdp.rename(columns = {"GeoName": "State"})
```

```python
gdp
```

```python
gdp.to_csv("state_gdp.csv", index = False)
```
